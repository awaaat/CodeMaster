#Tensor Multiplicaton

```python
import torch
tensor_A = torch.tensor([[1,2],
                        [4,5],
                        [7,8]])
tensor_B = torch.tensor([[9, 10, 11],
                        [12, 11, 14],
                        [14, 15, 16]])
print(torch.matmul(tensor_A.T, tensor_B))
print(tensor_A.shape, tensor_B.shape)
```
